```python
from os.path import dirname, join as pjoin
from scipy.io import wavfile
import scipy.io
```

```python
data_dir = pjoin(dirname(scipy.io.__file__),r'F:\myPolito\DSL\Final Project\dsl_data\audio\speakers\2BqVo8kVB2Skwgyb')
wav_fname = pjoin(data_dir, "0a3129c0-4474-11e9-a9a5-5dbec3b8816a.wav")
```

```python
samplerate , data = wavfile.read(wav_fname)
print(f"the data shape = {data.shape}")
print(f"the data rows = {data.shape[0]}")
length = data.shape[0]/samplerate
print(f"length = {length}s")
print(samplerate)
```

```python
import matplotlib.pyplot as plt
import numpy as np
time = np.linspace(0, length, data.shape[0])
plt.plot(time, data[:], label = "channel")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
fs = samplerate
powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(data, Fs = fs)
plt.xlabel("Time [s]")
plt.ylabel("Frequency(Hz)")
plt.show()

print(f"shape of time is: {time.shape}\n")
print(f"samples of time:\n {time[:5]}\n")

print(f"shape of freqenciesFound: {freqenciesFound.shape}")
print(f"samples of frequnciesFound:\n {freqenciesFound[:5]}\n")

print("Consider time as x_axis and Friquencies as y_axis powerspctrum is the number of times that frequency was repeated\n")
print(f"shape of powerspectrum is: {powerSpectrum.shape}\n")
print(f"samples of powerspectrum:\n {powerSpectrum[:5, :3]}\n")

```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
f, t, Zxx = signal.stft(data, fs = samplerate)
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()

print(f"shape of time is: {t.shape}\n")
print(f"samples of time:\n {t[:5]}\n")

print(f"shape of freqenciesFound: {f.shape}")
print(f"samples of frequnciesFound:\n {f[:5]}\n")

print("Consider time as x_axis and Friquencies as y_axis powerspctrum is the number of times that frequency was repeated\n")
print(f"shape of powerspectrum is: {Zxx.shape}\n")
print(f"samples of powerspectrum:\n {Zxx[:5, :3]}\n")
```
